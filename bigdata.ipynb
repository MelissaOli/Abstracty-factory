```python
import pandas as pd
import matplotlib.pyplot as plt

# Baixar os conjuntos de dados
!wget https://raw.githubusercontent.com/laderast/cvdNight1/master/data/fullPatientData.csv
!wget https://raw.githubusercontent.com/bonamicojedi/esp32/main/exported_data.csv

# Carregar os dados
data1 = pd.read_csv('fullPatientData.csv')
data2 = pd.read_csv('exported_data.csv')

# Dimensão dos conjuntos de dados
print("Dimensão do primeiro conjunto de dados:", data1.shape)
print("Dimensão do segundo conjunto de dados:", data2.shape)

# 20 primeiras linhas de cada conjunto de dados
print("\nPrimeiras 20 linhas do primeiro conjunto de dados:")
print(data1.head(20))
print("\nPrimeiras 20 linhas do segundo conjunto de dados:")
print(data2.head(20))

# Resumo estatístico dos conjuntos de dados
print("\nResumo estatístico do primeiro conjunto de dados:")
print(data1.describe())
print("\nResumo estatístico do segundo conjunto de dados:")
print(data2.describe())

# Média da idade agrupada por gênero no primeiro conjunto de dados
grupo_data1 = data1.groupby('gender')
idade_por_genero = grupo_data1['numAge'].mean()
print("\nMédia da idade por gênero no primeiro conjunto de dados:")
print(idade_por_genero)

# Gráfico de histograma para as calorias no segundo conjunto de dados
plt.hist(data2['Calories'])
plt.xlabel('Calorias')
plt.ylabel('Frequência')
plt.title('Distribuição de Calorias')
plt.show()
```
